```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("../data/processed/spotify_tracks_cleaned.csv")

#.head() shows first 5 rows
df.head()
```

```python
# gives us statistical summary of the data 
df.describe()
```

```python
# shows how many missing values there are in each column
df.isnull().sum()
```

```python
# tracks top 10 genres
df["track_genre"].value_counts().head(10)
```

```python
# esentially you group the trackgenre columns in groups
# then you take the mean of the popularity column for each group
# ex. pop has 100 songs, you group pop_genre into its own group
# and then you take the mean of the popularity of all 100 songs
# you sort the values from highest to lowest (since ascending is false)
# use head(10) to display top 10 genres with highest popularity

genre_popularity = df.groupby("track_genre")["popularity"].mean().sort_values(ascending=False)

genre_popularity.head(10)
```

```python
df["artists"].value_counts().head(10)
```

```python
df[["track_name", "artists", "track_genre", "popularity"]].sort_values(
    by="popularity", 
    ascending=False
).head(10)
```

```python
# top 10 genres by average popularity
top_genres = genre_popularity.head(10)

# chart with a wdith of 10 and a heigh of 6
plt.figure(figsize=(10, 6))

# creates a chart from top_genres
# kind = "bar" means "make a vertical bar chart"
# index -> x-axis
# values -> y-axis
top_genres.plot(kind="bar")

# title of chart
plt.title("Top 10 Genres by Average Popularity")

# x axis will be labeled as "genre"
plt.xlabel("Genre")

# y axis labeled as "average popularity"
plt.ylabel("Average Popularity")

# rotate the name of x axis labels by 45 degrees so no overlapping
plt.xticks(rotation=45)

# automatically adjusts spacing of graph
# labels, titles, and axis names will not be cut off
plt.tight_layout()

# display chart
plt.show()
```

# Check relationships between features
## Does danceability relate to popularity?

```python
plt.figure(figsize=(8, 6))

#alpha represents how transparent the dots are, from a scale of 0 to 1
plt.scatter(df["danceability"], df["popularity"], alpha = 0.3)

plt.title("Danceability vs Popularity")
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.tight_layout()
plt.show()
```

# Finding Correlations:
### Tells you whether two number columns move together.

```python
numeric_cols = [
    "popularity",
    "duration_minutes",
    "danceability",
    "energy",
    "loudness",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo"
]

# .corr() calculates correlation between every pair of columns in the dataframe
# values are from -1 to 1, where...
    # -1 means strong negative relationship
    # 0 means no relationship
    # 1 means strong positive relationship

# we use ["popularity"] to get the correlation of popularity with every other column

# using .sort_values(ascending=False), we get most positively correlated columns 
# at the top and most negatively correlated columns at the bottom

df[numeric_cols].corr()["popularity"].sort_values(ascending=False)
```
